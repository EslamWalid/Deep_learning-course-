```python
# importing libraries
from __future__ import print_function
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense,Flatten,Dropout
from keras.layers import Conv2D,Conv1D,MaxPool2D
from keras import backend as k 
```

```python

btch_size=128
num_class=10
epochs=12
```

```python
img_row,img_col=28,28
(trainx,trainy),(testx,testy) = mnist.load_data()
```

```python
# reshape the data
if k.image_data_format() == "channels_frist":
  trainx=trainx.reshape(trainx.reshape[0],1,img_row,img_col)
  testx=testx.reshape(testx.reshape[0],1,img_row,img_col)
  input_shap=(1,img_row,img_col)

else:
  trainx=trainx.reshape(trainx.shape[0],1,img_row,img_col)
  testx=testx.reshape(testx.shape[0],1,img_row,img_col)
  input_shap=(1,img_row,img_col)

# normalization 
trainx =trainx/255.0
testx =testx/255.0
print(trainx.shape)
 
```

```python
#covert lables (trainy) to binary because its categorical
tariny=keras.utils.np_utils.to_categorical(trainy,num_class)
testy=keras.utils.np_utils.to_categorical(testy,num_class) 
```

```python
from tensorflow.keras.optimizers import SGD
model=Sequential()
#input Layer
model.add(Conv2D(32,kernel_size=(3,3),activation="relu",input_shape=input_shap,padding='same'))
model.add(Conv2D(64,kernel_size=(3,3),activation="relu",padding='same'))
#pooling 
model.add(MaxPool2D(pool_size=(2,2),padding='same'))
#layers
#input layer
model.add(Flatten())
#hidden layer
model.add(Dense(128,activation='relu'))
#output layer
model.add(Dense(num_class,activation="softmax"))
opt=SGD(lr=0.01,momentum=0.9)
model.summary()
```

```python
import tensorflow as tf
model.compile(loss=tf.keras.losses.categorical_crossentropy,optimizer=tf.keras.optimizers.Adadelta(),metrics=['accuracy'])
model.fit(trainx,tariny,batch_size=32,epochs=5000,verbose=1,validation_data=(testx,testy))
```

```python
score=model.evaluate(testx,testy,verbose=0)
print('test loss= ',score[0],'\ntest accuracy= ',score[1])
```

```python
model.save('model.h5')
```
